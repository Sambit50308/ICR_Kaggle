```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
greeks = pd.read_csv('greeks.csv')
```

```python
y = data.Class
```

```python
data.head()
```

```python
metadata = greeks.iloc[:, [2,4]]
```

```python
train = data.iloc[:, 0:57]
```

```python
final_data = pd.concat([train, metadata], axis = 1).iloc[:, 1:60]
final_data.head()
```

```python
index = [0, 1, 12, 16, 17, 21, 29, 30, 31, 32, 34, 45, 47, 51, 56, 57] 
index
```

```python
new_data = final_data.iloc[:, index]
new_data.head()
```

```python
encoded_data = pd.concat([pd.get_dummies(new_data, columns = ['Beta', 'Delta'], drop_first = True), y], axis = 1)
rfm_data = encoded_data.dropna()
rfm_data.head()
```

```python
rfm_data.shape
```

```python
# Scaling the data

for i in range(14):
  rfm_data.iloc[:, i] = rfm_data.iloc[:, i] / sum(rfm_data.iloc[:, i])

rfm_data.head()
```

## SMOTE

```python
#pip install imbalanced-learn
```

```python
X = rfm_data.drop('Class', axis = 'columns')
Y = rfm_data['Class']
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(sampling_strategy = 'minority')
X_sm, y_sm = smote.fit_resample(X,Y)
```

```python
y_sm.value_counts()
```

```python
X_sm.head()
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.20, random_state = 15, stratify = y_sm)
```

```python
x_train.shape
```

## ANN

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(32, input_shape=(19,), activation='relu'),
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=150)
```

```python
yp = model.predict(x_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_pred_ann = np.array(y_pred)
```

```python
type(y_pred_ann)
```

```python
tp_ann = sum(y_pred_ann * y_test)
fn_ann = sum(y_test * (1 - y_pred_ann))

recall_ann = tp_ann/(tp_ann + fn_ann)
recall_ann
```
